## IHMM testing outcomes

```python
import numpy as np
from pymovements.events.detection.ihmm import ihmm
from pymovements.events.detection.idt import idt
from pymovements.events.detection.microsaccades import microsaccades
#from pymovements.gaze import from_numpy
from pymovements.synthetic import step_function
from pymovements.gaze.transforms_numpy import pos2vel

sampling_rate = 1000.0
```

### No fixations

```python
positions = step_function(
    length=6,
    steps=[1, 2, 3, 4, 5],
    values=[
        (100., 100.),
        (-100., -100.),
        (100., -100.),
        (-100., 100.),
        (200., 0.),
    ],
    start_value=(0., 0.),
)

velocities = pos2vel(positions,sampling_rate=sampling_rate)

ihmmEvents = ihmm(velocities,reestimation=True,name="ihmm_fix")

idtEvents = microsaccades(positions,name="idt_fix")
```

```python
#print(len(velocities))
ihmmEvents
```

```python
idtEvents
```
